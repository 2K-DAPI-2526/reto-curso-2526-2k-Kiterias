```python
import turtle

def estrella(puntas,tamaño):
    angulo=180-180/puntas
    for _ in range(puntas):
        turtle.forward(tamaño)
        turtle.right(angulo)

pantalla=turtle.Screen()
pantalla.setup(500,500)
pantalla.bgcolor("black")

turtle.speed(0)
turtle.pensize(3)
turtle.color("yellow")

puntas=int(input("Numero de puntas (mejor impar): "))

if puntas<=7:
    recomendado=180
elif puntas<=11:
    recomendado=140
else:
    recomendado=110

print("Tamaño recomendado:",recomendado)

tamaño=int(input("Introduce el tamaño: "))

if puntas%2==0:
    print("Aviso: para que la estrella cierre bien usa un numero impar")

estrella(puntas,tamaño)

turtle.done()
```
