```python
import random
filas=5
columnas=5
def generar_laberinto(muros):
    lab=[[' ' for _ in range(columnas)] for _ in range(filas)]
    for x,y in muros:
        lab[x][y]='X'
    lab[0][0]='E'
    lab[4][4]='S'
    return lab
def imprimir_laberinto(lab):
    print('+'+'---+'*columnas)
    for fila in lab:
        print('| '+' | '.join(fila)+' |')
        print('+'+'---+'*columnas)
def resolver_laberinto(lab):
    movimientos=[]
    visitado=[[False]*columnas for _ in range(filas)]
    direcciones={'Arriba':(-1,0),'Abajo':(1,0),'Izquierda':(0,-1),'Derecha':(0,1)}
    def dfs(x,y,camino):
        if x==filas-1 and y==columnas-1:
            movimientos.extend(camino)
            return True
        visitado[x][y]=True
        for mov,(dx,dy) in direcciones.items():
            nx,ny=x+dx,y+dy
            if 0<=nx<filas and 0<=ny<columnas:
                if not visitado[nx][ny] and lab[nx][ny]!='X':
                    if dfs(nx,ny,camino+[mov]):
                        return True
        visitado[x][y]=False
        return False
    dfs(0,0,[])
    return movimientos
mapas=[((0,1),(0,2),(0,3),(0,4),(1,1),(2,1),(2,3),(3,3),(4,0),(4,1),(4,2),(4,3)),
      ((0,1),(1,0),(1,1),(1,3),(2,2),(3,0),(3,2),(3,4),(4,1),(4,2)),
      ((0,1),(0,3),(1,1),(1,3),(2,0),(2,2),(3,2),(3,3),(4,0),(4,3))]
muros=random.choice(mapas)
lab=generar_laberinto(muros)
imprimir_laberinto(lab)
ruta=resolver_laberinto(lab)
print("Ruta:",ruta)
```
