```python
def inicial():
    return [["тЩЬ","тЩЮ","тЩЭ","тЩЫ","тЩЪ","тЩЭ","тЩЮ","тЩЬ"],["тЩЯ","тЩЯ","тЩЯ","тЩЯ","тЩЯ","тЩЯ","тЩЯ","тЩЯ"],["","","","","","","",""],["","","","","","","",""],["","","","","","","",""],["","","","","","","",""],["тЩЩ","тЩЩ","тЩЩ","тЩЩ","тЩЩ","тЩЩ","тЩЩ","тЩЩ"],["тЩЦ","тЩШ","тЩЧ","тЩХ","тЩФ","тЩЧ","тЩШ","тЩЦ"]]
def guardar(f,t):
    for r in t:
        f.write("\t".join(r)+"\n")
    f.write("\n")
def mostrar(t):
    for r in t:
        print(" ".join(c if c!="" else " " for c in r))
def copiar(t):
    return [r[:] for r in t]
nombre=input("Nombre fichero: ")
tablero=inicial()
with open(nombre,"w",encoding="utf-8") as f:
    guardar(f,tablero)
while True:
    opcion=input("Quieres hacer un movimiento (s/n): ")
    if opcion.lower()!="s":
        break
    print("Introduce la posicion de origen")
    filaorigen=int(input("Fila (1-8): "))-1
    colorigen=int(input("Columna (1-8): "))-1
    print("Introduce la posicion de destino")
    filadestino=int(input("Fila (1-8): "))-1
    coldestino=int(input("Columna (1-8): "))-1
    pieza=tablero[filaorigen][colorigen]
    tablero[filaorigen][colorigen]=""
    tablero[filadestino][coldestino]=pieza
    with open(nombre,"a",encoding="utf-8") as f:
        guardar(f,tablero)
mov=int(input("Que movimiento quieres mostrar: "))
with open(nombre,"r",encoding="utf-8") as f:
    datos=f.read()
bloques=datos.strip().split("\n\n")
if mov<0 or mov>=len(bloques):
    print("Movimiento no existe")
else:
    lineas=bloques[mov].split("\n")
    ver=[l.split("\t") for l in lineas]
    mostrar(ver)
```
